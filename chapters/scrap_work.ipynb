```python
import numpy as np

import matplotlib.pyplot as plt

# Define the objective function
def objective(x, y):
    return -x * y

# Create a grid of points
x = np.linspace(0, 2, 400)
y = np.linspace(0, 2, 400)
X, Y = np.meshgrid(x, y)
Z = objective(X, Y)

# Plot the contours of the objective function
plt.contour(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(label='Objective function value')

# Highlight the feasible region
plt.fill_between([0, 1], 0, 1, color='grey', alpha=0.3)

# Set plot labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Feasible Region and Objective Function Contours')

# Show the plot
plt.savefig('01142025_feasible_region.png')
plt.show()
```
